```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

# Floodfill usage

Guide is from [^1] 


[^1]: https://learnopencv.com/filling-holes-in-an-image-using-opencv-python-c/

```python
# Read image.
im = cv.imread('./images/nickel.jpeg', cv.IMREAD_GRAYSCALE)
plt.imshow(im)
```

```python
# Threshold.
# Set values equal to or above 200 to 0.
# Set values below 220 to 255.
th, im_th = cv.threshold(im, 100, 255, cv.THRESH_BINARY_INV)
```

```python
plt.imshow(im_th, cmap='gray')
```

```python
# Copy the thresholded image.
im_floodfill = im_th.copy()
```

```python
# Mask used to flood filling.
# Notice the size needs to be 2 pixels more than the image.
h, w = im_th.shape[:2]
mask = np.zeros((h+2, w+2), np.uint8)
```

```python
# Floodfill from point (0, 0), since we know (0, 0) is definitely the background.
cv.floodFill(im_floodfill, mask, (0, 0), 255)

plt.imshow(im_floodfill)
```

```python
# Invert floodfilled image
im_floodfill_inv = cv.bitwise_not(im_floodfill)
plt.imshow(im_floodfill_inv)
```

```python
# Combine the two images to get the foreground.
im_out = im_th | im_floodfill_inv
plt.imshow(im_out, cmap='gray')
```
